1. Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз. Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.

В данном случае $p = 0.8$, значит $q = 1 - p = 0.2$. Воспользуемся формулой Бернулли для определения вероятности того, что стрелок попадет в цель 85 раз:
### $$p(k=85) = C^{85}_{100}\cdot p^{85} \cdot q^{15} = \frac{100!}{15!\cdot85!}\cdot(0.8)^{85}\cdot(0.2)^{15} \approx 0.0481$$

```python
print(factorial(100)/(factorial(15) * factorial(85))*((0.8)**85)*((0.2)**15))
```

2. Вероятность того, что лампочка перегорит в течение первого дня эксплуатации, равна 0.0004. В жилом комплексе после ремонта в один день включили 5000 новых лампочек. Какова вероятность, что ни одна из них не перегорит в первый день? Какова вероятность, что перегорят ровно две?

Воспользуемся в данном случае формулой Пуассона, в данном случае $m_1 = 0$, а $m_2=2$. Кроме того, зная количество лампочек, участвующих в данном эксперименте, а также вероятность каждой перегореть в течение первого дня эксплуатации, получаем, что $\lambda = np = 5000\cdot0.0004=2$. Тогда,
### $$p_0 \approx \frac{\lambda^{m_1}}{m_1!}e^{-\lambda} = \frac{2^0}{0!}e^{-2} \approx 0.1353$$
### $$p_2 \approx \frac{\lambda^{m_2}}{m_2!}e^{-\lambda} = \frac{2^2}{2!}e^{-2} \approx 0.2707$$

```python
print(e**(-2))
print(2*e**(-2))
```

3. Монету подбросили 144 раза. Какова вероятность, что орел выпадет ровно 70 раз?

В данном случае $p = 0.5$, значит $q = 1 - p = 0.5$. Воспользуемся формулой Бернулли для определения вероятности того, что орел выпадет ровно 70 раз:
### $$p(k=70) = C^{70}_{144}\cdot p^{70} \cdot q^{74} = \frac{144!}{74!\cdot70!}\cdot(0.5)^{70}\cdot(0.5)^{74} \approx 0.0628$$

```python
print(factorial(144)/(factorial(70) * factorial(74))*((0.5)**144))
```

Какова вероятность того, что все мячи белые? Какова вероятность того, что ровно два мяча белые? Какова вероятность того, что хотя бы один мяч белый?

Вероятность, что все мячи белые:
### $$p = \frac{7\cdot6\cdot9\cdot8}{10\cdot9\cdot11\cdot10} \approx 0.3055$$
Теперь определим вероятность, что ровно два мяча белые. Это возможно в трех случаях:
   - Из первого ящика достали 2 белых мяча, из второго ноль;
   - Из первого и второго ящиков достали по 1 белому мячу;
   - Из первого ящика не достали ни одного белого мяча, а из второго достали два.

Тогда
### $$p = p_1 + p_2 + p_3$$
### $$p_1 = \frac{7\cdot6\cdot2\cdot1}{10\cdot9\cdot11\cdot10} \approx 0.0085$$
### $$p_2 = \frac{C^{1}_{3}\cdot C^{1}_{7}}{C^{2}_{10}}\cdot\frac{C^{1}_{2}\cdot C^{1}_{9}}{C^{2}_{11}}=\frac{3\cdot7\cdot2\cdot9}{9\cdot5\cdot5\cdot11} \approx 0.1527$$
### $$p_3 = \frac{3\cdot2\cdot9\cdot8}{10\cdot9\cdot11\cdot10} \approx 0.0436$$
### $$p\approx 0.2048$$
Теперь определим вероятность, что хотя бы один мяч белый. Удобнее расмотреть это как отрицание случая, когда все мячи черные. То есть вероятность данного события равна:
### $$p = 1 - \frac{3\cdot2\cdot2\cdot1}{10\cdot9\cdot11\cdot10} \approx 0.9988$$
