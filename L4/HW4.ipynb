# Урок 4. Непрерывные случайные величины. Функция распределения и плотность распределения вероятностей. Равномерное и нормальное распределение. Центральная предельная теорема

1. Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800]. Найдите ее среднее значение и дисперсию.

### $$M(A) = \frac{200+800}{2} = 500$$
### $$D(A) = \frac{(800-200)^2}{12} = 30000$$

2. О случайной непрерывной равномерно распределенной величине B известно, что ее дисперсия равна 0.2. Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5? Если да, найдите ее.

Правая граница может быть рассчитана исходя из известной левой границы и дисперсии:
### $$D(B)=\frac{(b-0.5)^2}{12} = 0.2 \Rightarrow b-0.5 = \sqrt{2.4} \Rightarrow b = 0.5 + \sqrt{2.4} \approx 2.05$$
Среднее значение может быть определено в этом случае как полусумма границ распределения:
### $$M(B)=\frac{0.5+0.5+\sqrt{2.4}}{2} = 0.5 +0.5\cdot\sqrt{2.4} \approx 1.27$$

3. Непрерывная случайная величина X распределена нормально и задана плотностью распределения:
### $$f(x) = \frac{1}{4\sqrt{2\pi}}e^{-\frac{(x+2)^2}{32}}$$
Найдите:
- M(X)
- D(X)
- std(X) (среднее квадратичное отклонение)

Как известно плотность распределения непрерывной случайной величины с нормальным распределением может быть записана следующим образом:
### $$f(x) = \frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{(x-a)^2}{2\sigma^2}},$$
где $a$ - матожидание, а $\sigma^2$ - дисперсия. <br>
Тогда 
### $$M(X) = -2;$$
### $$D(X) = 16;$$
### $$std(X) = 4.$$


4. Рост взрослого населения города X имеет нормальное распределение. Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см. Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:
- а). больше 182 см
- б). больше 190 см
- в). от 166 см до 190 см
- г). от 166 см до 182 см
- д). от 158 см до 190 см
- е). не выше 150 см или не ниже 190 см
- ё). не выше 150 см или не ниже 198 см
- ж). ниже 166 см.

Будем пользоваться в данном случае переходом к стандартному нормальному распределению и табличными значениями стандартного нормального распределения:
- а. $P(z > \frac{182-174}{8}) = P(z>1) \approx 0.1587$;
- б. $P(z > \frac{190-174}{8}) = P(z>2) \approx 0.0228$;
- в. $P(\frac{166-174}{8}\leq z \leq \frac{190-174}{8}) = P(-1 \leq z \leq 2) = P(z\leq2) -  P(z\leq-1)\approx 0.8185$;
- г. $P(\frac{166-174}{8}\leq z \leq \frac{182-174}{8}) = P(-1 \leq z \leq 1) = P(z\leq1) -  P(z\leq-1)\approx 0.6826$;
- д. $P(\frac{158-174}{8}\leq z \leq \frac{190-174}{8}) = P(-2 \leq z \leq 2) = P(z\leq2) -  P(z\leq-2)\approx 0.9544$;
- е. $P(z \leq \frac{150-174}{8} | z \geq \frac{190-174}{8}) = P(z \leq -3 | z \geq 2) = P(z\leq-3) +  P(z\geq2)\approx 0.0241$;
- ё. $P(z \leq \frac{150-174}{8} | z \geq \frac{198-174}{8}) = P(z \leq -3 | z \geq 3) = P(z\leq-3) +  P(z\geq3)\approx 0.0026$;
- ж. $P(z < \frac{166-174}{8}) = P(z < -1 ) \approx 0.1587$.

5. На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?

Рост такого человека отклоняется от математического ожидания роста в популяции на $\frac{190-178}{\sqrt{25}} = 2.4$ сигмы.
