1. Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

```python
a = np.array([100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150])
print(f'Размер исходного массива: {len(a)}.')
```

```python
mean_value = np.sum(a)/len(a)
var_biased = np.sum([(x - mean_value)**2 for x in a])/len(a)
var_unbiased = np.sum([(x - mean_value)**2 for x in a])/(len(a)-1)
std_value_biased = np.sqrt(var_biased)
std_value_unbiased = np.sqrt(var_unbiased)
print(f'Среднее арифметическое выборки - {mean_value:.4f}')
print(f'Смещенная оценка дисперсии для данной выборки - {var_biased:.4f}.')
print(f'Среднее квадратичное отклонение выборки на основе смещенной оценки дисперсии - {std_value_biased:.4f}.')
print(f'Несмещенная оценка дисперсии для данной выборки - {var_unbiased:.4f}.')
print(f'Среднее квадратичное отклонение выборки на основе несмещенной оценки дисперсии - {std_value_unbiased:.4f}.')
```

3. На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: a). первым спортсменом б). вторым спортсменом в). третьим спортсменом.

Для определения указанных вероятностей необходимо воспользоваться формулой Байеса. Вероятность выстрела каждым из спортсменов одинакова и равна $P(B_1)=P(B_2)=P(B_3)=\frac{1}{3}$, так как никакой дополнительной информации на этот счет в задании нет. Если обозначить событие попадания в цель, как $A$, то по условию имеем $P(A|B_1)=0.9; P(A|B_2)=0.8; P(A|B_1)=0.6$. Тогда полная вероятность попадания в мишень может быть определена как $P(A) = \frac{1}{3}\cdot(0.9+0.8+0.6) \approx 0.7667$. <br>
Тогда,
### $$P(B_1|A)=\frac{P(B_1)\cdot P(A|B_1)}{P(A)} = \frac{\frac{1}{3}\cdot0.9}{\frac{1}{3}\cdot(0.9+0.8+0.6)} = \frac{0.9}{2.3} \approx 0.3913 $$
### $$P(B_2|A)=\frac{P(B_2)\cdot P(A|B_2)}{P(A)} = \frac{\frac{1}{3}\cdot0.8}{\frac{1}{3}\cdot(0.9+0.8+0.6)} = \frac{0.8}{2.3} \approx 0.3478 $$
### $$P(B_3|A)=\frac{P(B_3)\cdot P(A|B_3)}{P(A)} = \frac{\frac{1}{3}\cdot0.6}{\frac{1}{3}\cdot(0.9+0.8+0.6)} = \frac{0.6}{2.3} \approx 0.2609 $$

4. В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. Какова вероятность, что он учится: a). на факультете A б). на факультете B в). на факультете C?

В данном случае аналогично потребуется формула Байеса. Обозначим сдачу сессии, как событие $S$. Тогда по условию $P(A)=P(B)=\frac{1}{4}$, а $P(C)=\frac{1}{2}$. Также из условия следует, что $P(S|A)=0.8; P(S|B)=0.7; P(S|C)=0.9$. Вероятность сдачи сессии в данном случае можно определить по формуле полной вероятности: $P = \frac{1}{4}\cdot0.8+\frac{1}{4}\cdot0.7+\frac{1}{2}\cdot0.9=0.825$. <br>
Тогда,
### $$P(A|S) = \frac{P(A)\cdot P(S|A)}{P(S)}=\frac{0.25\cdot0.8}{0.825}\approx0.2424$$
### $$P(B|S) = \frac{P(B)\cdot P(S|B)}{P(S)}=\frac{0.25\cdot0.7}{0.825}\approx0.2121$$
### $$P(C|S) = \frac{P(C)\cdot P(S|C)}{P(S)}=\frac{0.5\cdot0.9}{0.825}\approx0.5455$$
