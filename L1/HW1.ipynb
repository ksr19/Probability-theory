# Урок 1. Случайные события. Условная вероятность. Формула Байеса. Независимые испытания

1. Из колоды в 52 карты извлекаются случайным образом 4 карты. a) Найти вероятность того, что все карты – крести. б) Найти вероятность, что среди 4-х карт окажется хотя бы один туз.

a) Среди 52 карт 13 имеют масть крести, то есть задача сводится к выбору 4 карт из 13:
### $$ p = \frac{C^{4}_{13}}{C^{4}_{52}} = \frac{13!}{4! \cdot 9!} \cdot \frac{4! \cdot 48!}{52!} = \frac {10\cdot11\cdot12\cdot13}{49\cdot50\cdot51\cdot52} \approx 0.0026$$

б) Нас интересуют случаи, когда среди 4 карт как минимум 1 туз:
### $$ p = \frac{C^{1}_{4}\cdot C^{3}_{48} + C^{2}_{4}\cdot C^{2}_{48} + C^{3}_{4}\cdot C^{1}_{48} +  C^{4}_{4}\cdot C^{0}_{48}}{C^{4}_{52}} = \frac{4\cdot C^{3}_{48} + 6 \cdot C^{2}_{48} + 4\cdot C^{1}_{48}+ 1}{C^{4}_{52}}$$
### $$p = \frac{4\cdot\frac{46\cdot47\cdot48}{2\cdot3}+6\cdot\frac{47\cdot48}{2}+4\cdot48 + 1}{C^{4}_{52}}=\frac{69184+6768+192+1}{270725}\approx0.2813$$

2. На входной двери подъезда установлен кодовый замок, содержащий десять кнопок с цифрами от 0 до 9. Код содержит три цифры, которые нужно нажать одновременно. Какова вероятность того, что человек, не знающий код, откроет дверь с первой попытки?

Код содержит три цифры, каждая находится в интервале от 0 до 9. Так как происходит одновременное нажатие на кнопки, то порядок не важен. Значит искомая вероятность может быть определена следующим образом:
### $$p = \frac{C^{3}_{3}}{C^{3}_{10}} = \frac{1\cdot3!\cdot7!}{10!} = \frac{2\cdot3}{8\cdot9\cdot10}\approx0.0083$$

3. В ящике имеется 15 деталей, из которых 9 окрашены. Рабочий случайным образом извлекает 3 детали. Какова вероятность того, что все извлеченные детали окрашены?

### $$p = \frac{C^{3}_{9}}{C^{3}_{15}} = \frac{9!}{3!\cdot6!}\cdot\frac{3!\cdot12!}{15!}=\frac{7\cdot8\cdot9}{13\cdot14\cdot15}\approx0.1846$$

4. В лотерее 100 билетов. Из них 2 выигрышных. Какова вероятность того, что 2 приобретенных билета окажутся выигрышными?

### $$p = \frac{C^{2}_{2}}{C^{2}_{100}} = \frac{1\cdot2\cdot98!}{100!}=\frac{2}{99\cdot100}\approx0.00020$$
